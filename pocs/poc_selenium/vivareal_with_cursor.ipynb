```python
import time
import bs4
import os
import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support import expected_conditions as EC
from apis import VivaRealApi
```

```python
driver = webdriver.Chrome()
url = "https://www.vivareal.com.br/aluguel/santa-catarina/florianopolis/"
driver.get(url)
driver.set_window_size(1366, 800)
page = 1    
directory = f'data/raw/vivareal/{datetime.datetime.now().date()}/'
file_name = f'page-{page}.html'

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

while True:
    try:
        file_path = os.path.join(f'data/raw/vivareal/{datetime.datetime.now().date()}/', f'page-{page}.html')
        html_content = driver.page_source
        soup = bs4.BeautifulSoup(html_content, features="html5lib")
        with open(f'{file_path}', 'w', encoding='utf-8') as file:
            file.write(soup.prettify())
        time.sleep(2)
        driver.execute_script("window.scrollTo(0,9750)")
        driver.find_element(By.CSS_SELECTOR, ".pagination__item:nth-child(9) > .js-change-page").click()
    except Exception as e:
        print(e)
        break
    else:
        page += 1
```

```python
viva = VivaRealApi(cidade='florianopolis')
folder_path = f'data/raw/vivareal/{datetime.datetime.now().date()}/'

for file_name in os.listdir(folder_path):
    if os.path.isfile(os.path.join(folder_path, file_name)):
        with open(f'{folder_path}/{file_name}', 'r', encoding='utf-8') as file:
            html_content = file.read()
        soup = bs4.BeautifulSoup(html_content, 'html.parser')
        listings = viva._extract_listings_from_soup(soup=soup)
        added_listings = 0
        for i in listings:
            formatted = viva._format_listing(listing = i)
            success = viva._append_formatted_listing(listing=formatted)
            if success == True:
                added_listings += 1
```
